# Gr치fica de $f(x)=x-3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def g(x):
   return x
# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
```

```python
def h(x):
  return -3**(-x)
```

```python
f(-5), f(1)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10, 10,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(10,10))
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.axhline(0, color="black")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title("Suma de Funciones")
plt.grid(color="pink")
plt.savefig("output.png")  #Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.show()
```
