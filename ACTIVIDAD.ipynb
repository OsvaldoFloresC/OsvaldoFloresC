```python
import numpy as np
import matplotlib.pyplot as plt

F = np.array([-5,1,9,25,55,105])
x = np.array([0,1,2,3,4,5])
```

```python
F
```

```python
type(F)
```

```python
#graficamos los datos
plt.scatter(x,F, color="orange", marker='+')
plt.title("Gr√°fica de los datos.")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color="black", linestyle="--")

plt.show()
```

```python
def f(x):
  return x**3 - 2*x**2 + 7*x -5

x1=np.linspace(0,5,100)
```

```python
#graficamos los datos
plt.plot(x1, f(x1))
plt.scatter(x,F, color="orange", marker='o')
plt.title("Gr√°fica de los datos.")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(color="black", linestyle="--")

plt.show()
```

```python
G=np.diff(f(x))
print(G)
```

```python
G2=np.diff(G)
print(G2)
```

```python
G3=np.diff(G2)
print(G3)
```
