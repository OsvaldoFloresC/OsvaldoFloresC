Polinomio de Maclaurin
Esbozo de $f(x)=e^{x/2}  sen(x/3)$

```python
import matplotlib.pyplot as plt
import numpy as np
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10,10,100)
```

```python
def f(x):
  return np.exp(x/2)*np.sin(x/3)
def g(x):
  return (1/3)*x
def h(x):
  return (1/3)*x+ (1/6)*x**2
def w(x):
  return (1/3)*x+ (1/6)*x**2+ (23/648)*x**3
```

```python
# Grafica las funciones.
plt.figure(figsize=(8,6))
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, w(x))
plt.axhline(0, color="black")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.axvline(0, color="black")
plt.xlim(-30, 30)
plt.ylim(-30, 30)
plt.title("Polinomio de Maclaurin")
plt.grid(color="pink")
plt.savefig("output.png")  #Guardar gr치fico como im치gen PNG.
# Mostrarlo.
plt.show()
```

Podemos notar que la derivadas de la funci칩n $f(x)=e^{x/2}  sen(x/3)$
