Esbozo de
$f(x)=1+1/2x-1/8x^2+1/16x^3-15/16x^4$ **texto en negrita**

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Polinomio de Taylor
def f(x):
  return (x+1)**(1/2)
def g(x):
  return 1+ (1/2)*x
def h(x):
  return 1+ (1/2)*x- (1/8)*(x**2)
def d(x):
  return 1+ (1/2)*x- (1/8)*(x**2)+ (1/16)*(x**3)
def j(x):
  return 1+ (1/2)*x- (1/8)*(x**2)+ (1/16)*(x**3)-  (15/16)*(x**4)
```

Al aproximar las raices al polinomio tenemos \\
$P_{1}(x)= 1+ (1/2)(-0.5)- (1/8)(-0.5)^2+ (1/16)(-0.5)^3$ \\
$P_{2}(x)= 1+ (1/2)(-0.25)- (1/8)(-0.25)^2+ (1/16)(-0.25)^3$ \\
$P_{3}(x)= 1+ (1/2)(-0.875)- (1/8)(-0.875)^2+ (1/16)(-0.875)^3$ \\
$P_{4}(x)= 1+ (1/2)(0.5)- (1/8)(0.5)^2+ (1/16)(0.5)^3$




Entonces $(P_{1},P_{2},P_{3},P_{4})=$

```python
d(-0.5),d(-0.25),d(-0.875),d(0.5)
```

Siendo \\
$P_{1}=0.7109375$ \\
$P_{2}=0.8662109375$ \\
$P_{3}=0.4249267578125$ \\
$P_{4}=1.2265625$ \\

con estos valores podemos obtener el margen de error de la aproximacion con una diferencia muy sencilla.


```python

# Valores del eje X que toma el gráfico.
x=np.linspace(-10, 10,100)
```

```python
# Grafica las funciones.
plt.figure(figsize=(10,10))
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.plot(x, d(x))
#plt.plot(x, j(x))
plt.axhline(0, color="black")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.title("Polinomio de Taylor")
plt.grid(color="pink")
plt.savefig("output.png")  #Guardar gráfico como imágen PNG.
# Mostrarlo.
plt.show()
```

En conclusión las derivadas del polinomio resultan ser funciones continuas de $f(x)= \sqrt(x+1)$ \\
Aproxime $\sqrt(0.5), \sqrt(0.75), \sqrt(0.125)  y \sqrt(1.5)$ en el polinomio de Taylor de tercer grado.
