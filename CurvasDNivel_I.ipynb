```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x, y) = 1 + x
def f(x, y):
    return 1 + x

# Crear una cuadrícula de puntos (x, y)
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Crear el mapa de curvas de nivel
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.title('Mapa de curvas de nivel de f(x, y) = 1 + x')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Crear el gráfico de planos
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Gráfico de planos de f(x, y) = 1 + x')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x, y) = sqrt(4 - x^2 - y^2)
def f(x, y):
    return np.sqrt(4 - x**2 - y**2)

# Crear una cuadrícula de puntos (x, y)
x = np.linspace(-2, 2, 400)
y = np.linspace(-2, 2, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Filtrar valores inválidos (debido a la raíz cuadrada de negativos)
Z = np.where(X**2 + Y**2 > 4, np.nan, Z)

# Crear el mapa de curvas de nivel
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.title('Mapa de curvas de nivel de f(x, y) = sqrt(4 - x^2 - y^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Crear el gráfico de planos
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Gráfico de planos de f(x, y) = sqrt(4 - x^2 - y^2)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función f(x, y) = |x| + |y|
def f(x, y):
    return np.abs(x) + np.abs(y)

# Crear una cuadrícula de puntos (x, y)
x = np.linspace(-10, 10, 400)
y = np.linspace(-10, 10, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Crear el mapa de curvas de nivel
plt.contour(X, Y, Z, levels=20, cmap='viridis')
plt.colorbar()
plt.title('Mapa de curvas de nivel de f(x, y) = |x| + |y|')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Crear el gráfico de planos
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')
ax.set_title('Gráfico de planos de f(x, y) = |x| + |y|')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```
